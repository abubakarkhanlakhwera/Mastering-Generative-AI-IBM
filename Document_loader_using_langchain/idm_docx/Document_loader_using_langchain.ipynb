- Useful for **AI chatbots, document search, and knowledge retrieval**.

- Extracts **embedded images, tables, and annotations**.
- Unlike `PyPDF2`, which mainly extracts raw text.

|--------|-------------|

| `soup.find_all("tag")` | Finds all occurrences of a tag (e.g., `soup.find_all("a")`) |
| `soup.get_text()` | Extracts text content from the HTML |
| `soup.find("tag")["attribute"]` | Extracts an attribute (e.g., `soup.find("a")["href"]`) |
